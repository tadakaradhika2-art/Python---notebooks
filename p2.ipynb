```python
sentence="python is easy python is easy"
freq={}
x=sentence.split()
for i in x:
   if i in freq:
       freq[i]+=1
   else:
       freq[i]=1
print (freq )      
```

```python
p1={"a":10,"b":20,"c":30}
p2={ "b":20,"d":5}
x=p1.copy()
for key,value in p2.items():
    if key in x:
       x[key]+=value
    else:
        x[key]=value

print("merged dict:",x)        
        
        
```
